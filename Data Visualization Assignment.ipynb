```python
import pandas as pd
import seaborn  as sns
import matplotlib.pyplot as plt
```

```python
def load_dataset (Health_heart_experimental):
  return pd.read_csv(Health_heart_experimental)
```

```python
df = load_dataset("Health_heart_experimental.csv")
df.head(10)
```

## Line Chart

```python
BMI = [20,21,21,32]
SysBP = [106,109,112,141]

# create a line chart, BMI on x-axis, SysBP on y-axis
plt.plot(SysBP, BMI, color='green', marker='+', linestyle='dotted')

# add a title
plt.title("SysBp vs BMI")

# add a label to the y-axisYears
plt.ylabel("SysBP")
plt.xlabel("BMI")
plt.show()
```

From this line graph, we can observe that people with higher BMI generally have higher SysBP(systolic blood pressure).

## Bar Chart

```python
sns.countplot(data=df, x='sex')
df['heightCm'].value_counts(5)
```

From above bar graph, we can find gender 0 are taller than gender 1.

```python
sns.barplot(x='indication',y='age',hue='sex',data=df)
```

We can see that people of gender 1 have more indications of both 0 and 1.

## Pie Chart

```python
df['sex'].value_counts().plot(kind='pie',autopct='%0.2f')
```

In this pie chart, both sexes have almost same number of people, gender 0 having 50.09% and gender 1 having 49.91% of the total number of people.

## Box Plot

```python
sns.boxplot(df['age'])
```

We can observe that the age of most of the people in between 30 and 60.

```python
sns.boxplot(x='sex',y='age',hue='indication',data=df)
```

People with both sexes have nearly the same age group  with indications 0 and 1.

## Histogram

```python
plt.hist(df['weightKg'])
```

This histogram has skew right nature.

```python
plt.hist(df['heightCm'])
```

This histogram is of skew left nature.

```python
sns.kdeplot(df[df['BMI'] >20]['age'], label='overweight')
sns.kdeplot(df[df['BMI'] ==20]['age'], label='healthy')
sns.kdeplot(df[df['BMI'] <20]['age'], label='underweight')

plt.title('BMI')
plt.xlabel('age')
plt.ylabel('Density')
plt.legend()
plt.show()
```

The people of age 30-40 are mostly healthy and least underweight. The people of age 50-60 are mostly underweight and least healthy. 

## Scatter Plot

```python
sns.scatterplot(x='heightCm', y='weightKg',hue='sex', style='indication',data=df)
plt.show()
```

This plot shows that the weight generally increases along the increase in height.
